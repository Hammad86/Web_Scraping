```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
import time
import csv
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
with open('sampleData4.csv', 'a+', encoding='utf-8-sig', newline='') as f:
    writer = csv.writer(f)
    for i in range(4001,5194):
        try:
            driver.get('https://www.iosweb.net/findortho.aspx')
            id_input = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//input[@id='MainContent_txt_RegNo'])"))
            )
            search_btn = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//input[@id='MainContent_bnt_Search'])"))
            )
            id_input.send_keys(f"LM{i}")
            search_btn.click()

            driver.execute_script("window.scrollTo(0, 500)") 

            name = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//span[@id='MainContent_lblName'])"))
            )

            registration_id = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//span[@id='MainContent_lbl_RegNo'])"))
            )

            phone_number = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//span[@id='MainContent_lblMobile'])"))
            )

            email = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//span[@id='MainContent_lblEmail'])"))
            )
            img = WebDriverWait(driver, 10).until(
                            EC.presence_of_element_located((By.XPATH, f"(//img[@id='MainContent_Img_Member'])"))
            )
        
            print(name.text)
            print(registration_id.text)
            print(phone_number.text)
            print(email.text)
            print(img.get_attribute('src'))
            writer.writerow([name.text,registration_id.text,phone_number.text,email.text,img.get_attribute('src')])
        except:
            print(f"error {i}")
```
