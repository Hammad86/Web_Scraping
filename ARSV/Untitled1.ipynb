```python
import requests
import csv
import time

api = 'https://servicios.dgac.gob.cl/arsv-web/rest/solicitud/publico/buscarSolicitudArsv/'
end = '/espanol'
req_type = 'NIL'
req_date = 'NIL'
p_name = 'NIL'
l_name = 'NIL'
address = 'NIL'
email = "NIL"
registration = 'NIL'
purpose = 'NIL'
call_Sign = 'NIL'
f_name = 'NIL'
f_mail = 'NIL'
f1 = open("asr_data_3.csv", 'a+', encoding='utf-8-sig', newline='')
csv_f1 = csv.writer(f1)
for i in range(20278,21000):
    try:
        response = requests.get(f"{api}{i}{end}")
        print(response)
        response = response.json()
        url = f"https://servicios.dgac.gob.cl/arsv-web/solicitudes.html#/verSolicitudArsv/{i}"
        data = response['data']
        req_type = data['tipoSolicitud']['nombre']
        req_date = data['fechaCreacion']
        p_name = data['nombreEmpresa']
        l_name = data['razonSocial']
        address = data['direccionFacturacion']
        email = data['correoElectronico']
        registration = data['matricula']
        purpose = data['objetoVuelo']['nombre']
        call_Sign = data['callSign']
        f_name = data['embarcador']
        f_mail = data['correoElectronicoOperador']
        csv_f1.writerow([url,req_type,req_date,p_name,l_name,address,email,registration,purpose,call_Sign,f_name,f_mail])
        print(i)
    except:
        print('error')
        continue
```
