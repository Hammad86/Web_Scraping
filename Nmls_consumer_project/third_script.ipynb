```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.remote.webelement import WebElement
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv
import names
from twocaptcha import TwoCaptcha
import sys
import os
```

```python
#script 1 for getting name and number with captcha
with open("nameLinks2.csv") as f:
    data = csv.reader(f)
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.maximize_window()
    for i in data:
        for j in i:
            print(j)    
            try:
                driver.get(j)
                element = WebDriverWait(driver, 5).until(
                        EC.element_to_be_clickable((By.XPATH, "//p[@class='individual']")) )
                number = WebDriverWait(driver, 5).until(
                        EC.element_to_be_clickable((By.XPATH, "//td[@class='divider'][2]")) )
                
                with open('name&number2.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([element.text,number.text])
                print(element.text)
                print(number.text)
            except:
                print('break')
            try:
                image = WebDriverWait(driver, 5).until(
                                        EC.presence_of_element_located((By.XPATH, "//img[@class='LBD_CaptchaImage']")) )
#                 img_src = image.get_attribute('src')
                image.screenshot('captchas/2.png')
                api_key = os.getenv('APIKEY_2CAPTCHA', '50456618cb7bd95242302e630316801a')

                solver = TwoCaptcha(api_key)
                result = solver.normal('captchas/2.png')
                input_tag = driver.find_element_by_xpath("//input[@id='ctl00_MainContent_txtTuringText']")
                print(result['code'])
                input_tag.send_keys(result['code'])
                click = WebDriverWait(driver, 5).until(
                                    EC.presence_of_element_located((By.XPATH, "//input[@id='ctl00_MainContent_btnContinue']")) )
                click.click()
                element = WebDriverWait(driver, 5).until(
                            EC.element_to_be_clickable((By.XPATH, "//p[@class='individual']")) )
                number = WebDriverWait(driver, 5).until(
                        EC.element_to_be_clickable((By.XPATH, "//td[@class='divider'][2]")) )

                with open('bhund.csv', 'a+', encoding='utf-8-sig', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow([element.text,number.text])
                print(element.text)
                print(number.text)
            except:
                print('break2')
                continue
            
```

```python
#script 2 for getting links with captcha
driver = webdriver.Chrome(ChromeDriverManager().install())
with open('newLinksData.csv', 'a+', encoding='utf-8-sig', newline='') as f:
    writer = csv.writer(f)
    flag = True
    while flag:
        try:
            driver.get('https://nmlsconsumeraccess.org/')
            search_input = WebDriverWait(driver, 5).until(
                                EC.presence_of_element_located((By.XPATH, "//input[@class='swap_value field']")) )
            a = names.get_first_name()
            b = names.get_last_name()
            search_input.send_keys(f"{a} {b}")
            search_btn = WebDriverWait(driver, 5).until(
                                    EC.presence_of_element_located((By.XPATH, "//input[@id='searchButton']")) )
            search_btn.click()
        except:
            continue
        
        try:
            image = WebDriverWait(driver, 5).until(
                                    EC.presence_of_element_located((By.XPATH, "//img[@class='LBD_CaptchaImage']")) )
#                 img_src = image.get_attribute('src')
            image.screenshot('captchas/2.png')
            api_key = os.getenv('APIKEY_2CAPTCHA', '50456618cb7bd95242302e630316801a')

            solver = TwoCaptcha(api_key)
            result = solver.normal('captchas/2.png')
            input_tag = driver.find_element_by_xpath("//input[@id='ctl00_MainContent_txtTuringText']")
            print(result['code'])
            input_tag.send_keys(result['code'])
            click = WebDriverWait(driver, 5).until(
                                EC.presence_of_element_located((By.XPATH, "//input[@id='ctl00_MainContent_btnContinue']")) )
            click.click()
            
            flag2 = True
            while flag2:
                try:
                    headers = WebDriverWait(driver, 30).until(
                                        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='individual']")) )
                    for i in headers:
                        a = i.get_attribute('onclick')
                        first = a.index("(")
                        last = a.index(")")
                        last = last - 1
                        first = first + 2
                        link=f"https://nmlsconsumeraccess.org/{a[first:last]}"
                        print(link)
                        writer.writerow([link])
                    time.sleep(2)
                    nextBtn = driver.find_element_by_xpath("//li[@class='nextOn']")
                    time.sleep(2)
                    nextBtn.click()
                except:
                    print("stop1!!!")
                    flag2 = False
            
        except:
            print('break1')
            flag2 = True
            while flag2:
                try:
                    headers = WebDriverWait(driver, 5).until(
                                        EC.presence_of_all_elements_located((By.XPATH, "//a[@class='individual']")) )
                    for i in headers:
                        a = i.get_attribute('onclick')
                        first = a.index("(")
                        last = a.index(")")
                        last = last - 1
                        first = first + 2
                        link=f"https://nmlsconsumeraccess.org/{a[first:last]}"
                        print(link)
                        writer.writerow([link])
                    time.sleep(2)
                    nextBtn = driver.find_element_by_xpath("//li[@class='nextOn']")
                    time.sleep(2)
                    nextBtn.click()
                except:
                    print("stop!!!")
                    flag2 = False
            continue
```
